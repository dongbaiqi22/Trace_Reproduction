```python
import numpy as np
```

```python
# load data

data = np.load("./", allow_pickle=True)

emb = data['emb']
coords = data['coords']
labels = data['labels']
```

```python
# Visualization: latent space (u_train)
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(emb[:, 0], emb[:, 1], s=10, c=labels, cmap="Set1", alpha=0.3)
plt.gca().set_aspect("equal", adjustable="box")
plt.title(f"Neurons in latent space")
plt.xlabel("u[:,0]")
plt.ylabel("u[:,1]")
plt.show()
```

```python
# Compute Silhouette Score
from sklearn.metrics import silhouette_score

score = silhouette_score(emb, labels)
print(f"Silhouette Score: {score:.4f}")
```
